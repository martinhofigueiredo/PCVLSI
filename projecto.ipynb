# Multi-stage interconnection network

## Trying to simulate the logic of this circuit


```python
#Brute Force (?)

#      ______
#     |      |
# X0--|      |--A
#     |      |  
# X1--|      |--B
#     |______|
# 
def benes1(tupple, sel):
    (a, b) = tupple
    return (b, a) if sel else (a, b)


print(benes1(("a", "b"), 0))
print(benes1(("a", "b"), 1))
```

```python
def benes2(X, S):
    (b0A, b0B) = benes1((X[0],X[1]),S[0])
    (b1A, b1B) = benes1((X[2],X[3]),S[1])
    
    #print(f"Stage 1 - {(b0A, b0B)} and {(b1A, b1B)} ")

    (b2A, b2B) = benes1((b0A,b1A),S[2])
    (b3A, b3B) = benes1((b0B,b1B),S[3])
    
    #print(f"Stage 2 - {(b2A, b2B)} and {(b3A, b3B)} ")

    (b4A, b4B) = benes1((b2A,b3A),S[4])
    (b5A, b5B) = benes1((b2B,b3B),S[5])
    
    #print(f"Stage 3 - {(b4A, b4B)} and {(b5A, b5B)} ")

    return [b4A, b4B, b5A, b5B]
```

```python

X = ["0","1","2","3"]
#X = ["A","B","C","D"]

S = [0,0,0,0,0,0]
Y = []

print(f"|S dec\t|   S bin \t|[Y0 , Y1 , Y2 , Y3 ]|")
print(f"|-----\t|-----------\t|--------------------|")

for i in range(0,2**6):
    S = [int(digit) for digit in str(format(i, '06b'))]
    Y.append(benes2(X, S))
    print(f"| {i}\t|  0b{format(i, '06b')}\t|{benes2(X, S)}|")
```

```python
aux = []
for item in Y:    
    aux.append(item[0]+item[1]+item[2]+item[3])

words = aux
word_count = {}
word_positions = {}

for i, word in enumerate(words):
    if word in word_count:
        word_count[word] += 1
        word_positions[word].append(i)
    else:
        word_count[word] = 1
        word_positions[word] = [i]

same_count = 0
different_count = 0

for word, count in word_count.items():
    if count > 1:
        same_count += count
    else:
        different_count += 1

unique_count = len(word_count)

print("Same words count:", same_count)
print("Different words count:", different_count)
print("Unique words count:", unique_count)
print("Positions of same words:")
i=1
for word, positions in word_positions.items():
    if len(positions) > 1:
        print(f"{i}:\t",word, ":", positions)
        i=i+1
```
